### Converting our coordinate reference system

We need to first get our `xcoord` and `ycoord` fields associated with a coordinate reference system so that they can be transformed into the coordinate system that ploty is expecting.

The CERF data is produced using the USA Contiguous Albers Equal Area Conic projected coordinate system in meters. GeoPandas allows you to set these once a geometry field has been made.  

We need to build our geometry field from our CERF coordinate values using:
`gpd.points_from_xy(new_df["xcoord"], new_df["ycoord"])`

We are also telling the geopandas dataframe that we know that data is in USA Contiguous Albers Equal Area Conic and we are going to give it a shortcut known as an EPSG code to get that set up for us.  You can see more about the EPSG code if you are interested here:  https://epsg.io/102003


```python
gdf = gpd.GeoDataFrame(new_df, 
                       geometry=gpd.points_from_xy(new_df["xcoord"], new_df["ycoord"]),
                       crs="ESRI:102003")

# let's take a look at our geometry field
gdf.geometry

```

Our next step is to transform, or reproject, our coordinate reference system from CERF's native projection to a geographic coordinate system (one that has latitude and longitude values rather than x, y coordinates) that Plotly expectes.  

Again, we use our handy EPSG code to tell it we want the data in WGS 84 (https://epsg.io/4326). 

We use the `inplace` setting meaning that we don't want to create a new variable, we just want to modify what we have "inplace"


```python
gdf.to_crs("EPSG:4326", inplace=True)

# notice the change in the geometry of each point to a lat, lon value
gdf.geometry
```
